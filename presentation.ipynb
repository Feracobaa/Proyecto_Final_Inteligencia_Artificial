## 1. Visión General del Proyecto

El proyecto implementa una plataforma web interactiva para demostrar conceptos de **Regresión Lineal**, **Regresión Logística** y **Clasificación Supervisada**.

* Simulador 2D de regresión lineal con entrenamiento por mínimos cuadrados (OLS) y gradiente descendente (GD).
* Clasificador binario basado en regresión logística multivariable.
* Herramientas de carga de datasets CSV, visualización de métricas y curva ROC.

La lógica está implementada puramente en **HTML**, **CSS** y **JavaScript**, sin backend, facilitando su despliegue en entornos estáticos como GitHub Pages.

## 4. Fragmentos Clave del Código

A continuación se muestra un extracto del módulo de entrenamiento por **Mínimos Cuadrados (OLS)** en `app.js`:

```javascript
function fitOLS() {
    const N = points.length;
    let sumX = 0, sumY = 0;
    for (let p of points) {
        sumX += p.x;
        sumY += p.y;
    }
    const meanX = sumX / N;
    const meanY = sumY / N;
    let num = 0, den = 0;
    for (let p of points) {
        num += (p.x - meanX) * (p.y - meanY);
        const xDiff = p.x - meanX;
        den += xDiff * xDiff;
    }
    if (den === 0) {
        modelParams.m = 0.0;
        modelParams.b = meanY;
    } else {
        modelParams.m = num / den;
        modelParams.b = meanY - modelParams.m * meanX;
    }
    updateParamsDisplay();
    updateSimulatorMetrics();
}
```
